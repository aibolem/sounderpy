# **PART B** - Getting Data

* **Step 1:** Determine a few parameters
  - parameters, or 'arguments', point SounderPy towards what data you want to access
  - These include the model name, site ID, forecast hour, etc.
* **Step 2:** Load the data and store it in a variable

-

----

-

**Before we get started**, lets investigate the data SounderPy can get:

* **Latest Model Forecasts**: HRRR, RAP, NAM-NEST, NAM, SREF, HIRESW, GFS
* **Archived Model Forecasts**: HRRR, RAP, NAM, GFS

Next, we have to discuss 'BUFKIT':

* BUFKIT model data is forecast model data for specific 'site IDs'.
* SounderPy can *only* get forecast model data for valid BUFKIT sites
* An example of a valid site ID is 'KMOP', 'KLAN', or 'DTX'
* [Here is a full map of station IDs from PSU](https://http://www.meteo.psu.edu/bufkit/CONUS_NAMNEST_00.html).

-

-----

-

**To get model forecast data from SounderPy**, we will use this line of code:

``model_data = spy.get_bufkit_data(model_name, site_id, forecast_hour)``

*   ``model_data`` is the name we store the data in (this can be whatever you want!)
*   ``spy`` is calling up the SounderPy software saying "hey, I'd like to access one of SounderPy's tools"
*   ``get_bufkit_data()`` is telling SounderPy what tool we want to use
*  ``model_name``, ``site_id``, & ``forecast_hour`` are all parameters that the ``get_bufkit_data`` tool needs to get us the data.

-

*Using the code described above, we can get the latest HRRR forecast, at forecast hour 0, for Mount Pleasant!*


```python
# STEP 1 - Determine a few parameters
# Declare a BUFKIT site ID, forecast model, & forecast hour

site_id = 'kmop'
model_name = 'hrrr'
forecast_hour = 6

# ===================================================================== #

# STEP 2 - Load the data and store it in a variable
# We add our parameters defined above into the tool below
model_data = spy.get_bufkit_data(model_name, site_id, forecast_hour)
```

...and there we go, we now have forecast model data stored in the ``model_data`` variable. Next we can use it to make a sounding plot.

-

# **PART C** - Plotting the data

* **Step 1:** Simply add our ``model_data`` variable to SounderPy's plot tool
* **Step 2:** (Optional) Customize the plot!

-

----

**To plot the data on a Sounding**, we will use this line of code:

``spy.build_sounding(model_data)``

*   ``spy`` is calling up the SounderPy software saying "hey, I'd like to access one of SounderPy's tools"
*   ``build_sounding()`` is telling SounderPy that we want the tool that plots soundings
*   ``model_data`` is the variable we stored the data in, so we just 'give it to' the ``build_sounding()`` tool!

-

------

-

**Optionally, you can customize** the sounding plots with a few simple settings:

* **Style**: choose between ``'full'`` or ``'simple'``
* **Dark Mode:** use ``dark_mode=True``
* **Color-Blind Friendly Settings:** use ``color_blind=True``
* **Hodograph Only Plot:** just change ``build_sounding()`` to ``build_hodograph()``
  - *Note: the hodograph does have `dark_mode` but does not have the `color_blind` setting.*
  - An additional setting for hodographs only transforms the orientation to **Storm-Relative**, just set ``sr_hodo=True``

**An example of the code using settings**:
* ``spy.build_sounding(model_data, dark_mode=True, color_blind=True)``
* ``spy.build_hodograph(model_data, dark_mode=True)``

-

Its as simple as that!


```python
# STEP 1 - Simply add our `model_data` variable to SounderPy's plot tool
# we can also declare a few settings and add them to the tool.

style       = 'full' # try changing to... 'simple'
dark_mode   = False  # try changing to.... True
color_blind = False  # try changing to.... True

spy.build_sounding(model_data, style, color_blind, dark_mode)

# you can 'right-click' the image below to open it in a new tab or to save it
# to your device!
```
